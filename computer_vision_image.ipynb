```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image_path='/content/dog.jpeg'
dog_image=cv2.imread(image_path)

if dog_image is not None:
  print("Original Image:")
  cv2_imshow(dog_image)

  print("Grayscale Image:")
  gray_image=cv2.cvtColor(dog_image,cv2.COLOR_BGR2GRAY)
  cv2_imshow(gray_image)

  print("Blurred based on median filter:")
  median_blur=cv2.medianBlur(dog_image,5)
  cv2_imshow(median_blur)

  print("Blurred based on gaussian filter:")
  gaussian_blur=cv2.GaussianBlur(dog_image,(5,5),0)
  cv2_imshow(gaussian_blur)

  print("Blurred based on bilateral filter:")
  bilateral_blur=cv2.bilateralFilter(dog_image,9,75,75)
  cv2_imshow(bilateral_blur)

  print("Blurred based on mean filter")
  mean_blur=cv2.blur(dog_image,(5,5))
  cv2_imshow(mean_blur)

  print("Sobel Edge Detection:")
  sobel_x=cv2.Sobel(dog_image,cv2.CV_64F, 1, 0, ksize=5)
  sobel_y=cv2.Sobel(dog_image,cv2.CV_64F, 0, 1, ksize=5)
  combined_sobel=cv2.convertScaleAbs(sobel_x)+cv2.convertScaleAbs(sobel_y)
  print(">> sobel_x")
  cv2_imshow(sobel_x)
  print(">> sobel_y")
  cv2_imshow(sobel_y)
  print(">> combined_sobel")
  cv2_imshow(combined_sobel)

  print("Canny Edge Detection:")
  canny_edges=cv2.Canny(dog_image,100,200)
  cv2_imshow(canny_edges)

  print("Laplacian Edge Detection:")
  laplacian=cv2.Laplacian(dog_image,cv2.CV_64F)
  cv2_imshow(laplacian)
else:
  print("Failed to load the image")
```
